```python
def mon_produit_scalaire(a, b): 
    """ 
    Calcule le produit scalaire de deux vecteurs a et b. 
    Args: 
        a, b : tableaux NumPy 1-D de même dimension 
    Returns: 
        x : scalaire (float) 
    """ 
    x = 0 
    for i in range(a.shape[0]): 
        x = x + a[i] * b[i]  
    return x 
  
 # Test 
a = np.array([1, 2, 3, 4]) 
b = np.array([-1, 4, 3, 2]) 
print(f"Mon résultat   : {mon_produit_scalaire(a, b)}") 
print(f"NumPy np.dot() : {np.dot(a, b)}")   # doit être 24 

```

```python
X_train = np.array([[2104, 5, 1, 45], 
                    [1416, 3, 2, 40], 
                    [ 852, 2, 1, 35]]) 
y_train = np.array([460, 232, 178]) 
print(f"Shape de X_train : {X_train.shape}") 
print(f"Shape de y_train : {y_train.shape}") 
print(X_train) 
print(y_train) 
```

```python
b_init = 785.1811367994083 
w_init = np.array([0.39133535, 18.75376741, -53.36032453, -26.42131618]) 
  
print(f"Shape de w_init : {w_init.shape}") 
print(f"Type  de b_init : {type(b_init)}") 
```

```python

def predict_single_loop(x, w, b): 
    """ 
    Prédiction par régression linéaire (version boucle). 
    Args: 
        x (ndarray, shape (n,)) : un exemple avec n features 
        w (ndarray, shape (n,)) : paramètres du modèle 
        b (scalar)              : biais du modèle 
    Returns: 
        p (scalar) : valeur prédite 
    """ 
    n = x.shape[0] 
    p = 0 
    for i in range(n): 
        p = p + x[i] * w[i] 
    p = p + b 
    return p 
  
# Test 
x_vec = X_train[0, :] 
print(f"x_vec shape : {x_vec.shape}, valeur : {x_vec}") 
f_wb = predict_single_loop(x_vec, w_init, b_init) 
print(f"Prediction  : {f_wb}") 
```

```python
def predict(x, w, b): 
    """ 
    Prédiction par régression linéaire (version vectorisée). 
    Args: 
        x (ndarray, shape (n,)) : un exemple avec n features 
        w (ndarray, shape (n,)) : paramètres du modèle 
        b (scalar)              : biais du modèle 
    Returns: 
        p (scalar) : valeur prédite 
    """ 
    p = np.dot(x, w) + b 
    return p 
  
x_vec = X_train[0, :] 
f_wb = predict(x_vec, w_init, b_init) 
print(f"Prediction (vectorisée) : {f_wb}")
```

```python
for i in range(X_train.shape[0]): 
    prediction = predict(X_train[i], w_init, b_init) 
    print(f"Exemple {i} — Predit : {prediction:.2f}, Reel : {y_train[i]}") 
```

```python
def compute_cost(X, y, w, b): 
    """ 
    Calcule la fonction de cout pour la régression linéaire multiple. 
    Args: 
        X (ndarray, shape (m,n)) : donnees d'entrainement 
        y (ndarray, shape (m,))  : valeurs cibles 
        w (ndarray, shape (n,))  : parametres du modele 
        b (scalar)               : biais 
    Returns: 
        cost (scalar) : valeur du cout 
    """ 
    m = X.shape[0] 
    cost = 0.0 
    for i in range(m): 
        f_wb_i = np.dot(X[i], w) + b       # prédiction pour l'exemple i 
        cost = cost + (f_wb_i - y[i])**2   # erreur quadratique 
    cost = cost / (2 * m) 
    return cost 
  
# Test avec les paramètres optimaux 
cost = compute_cost(X_train, y_train, w_init, b_init) 
print(f"Cout avec w_init, b_init : {cost}") 

```

```python
w_zeros = np.zeros(4) 
b_zero  = 0. 
cost_zero = compute_cost(X_train, y_train, w_zeros,b_zero ) 
print(f"Cout avec w=0, b=0 : {cost_zero:.2f}")
```

```python
def compute_gradient(X, y, w, b): 
    """ 
    Calcule le gradient de la fonction de cout. 
    Args: 
        X (ndarray, shape (m,n)) : données d'entrainement 
        y (ndarray, shape (m,))  : valeurs cibles 
        w (ndarray, shape (n,))  : paramètres du modèle 
        b (scalar)               : biais 
    Returns: 
        dj_dw (ndarray, shape (n,)) : gradient par rapport a w 
        dj_db (scalar)              : gradient par rapport a b 
    """ 
    m, n = X.shape 
    dj_dw = np.zeros((n,)) 
    dj_db = 0. 
  
    for i in range(m): 
        err = (np.dot(X[i], w) + b) - y[i] 
        for j in range(n): 
            dj_dw[j] = dj_dw[j] + err * X[i, j] 
        dj_db = dj_db + err 
  
    dj_dw = dj_dw / m 
    dj_db = dj_db / m 
    return dj_db, dj_dw 
  
# Test 
tmp_dj_db, tmp_dj_dw = compute_gradient(X_train, y_train, w_init, b_init) 
print(f"dj_db : {tmp_dj_db}") 
print(f"dj_dw : {tmp_dj_dw}") 
```

```python
def gradient_descent(X, y, w_in, b_in, cost_function,gradient_function, alpha, num_iters): 
 
        """  
        Exécute la descente de gradient par batch.  
        Args:  
           X (ndarray, shape (m,n)) : données d'entraînement  
           y (ndarray, shape (m,)) : valeurs cibles  
           w_in (ndarray, shape(n,)): paramètres initiaux  
           b_in (scalar) : biais initial  
           cost_function : fonction de coût  
           gradient_function : fonction de gradient  
           alpha (float) : taux d'apprentissage  
           num_iters (int) : nombre d'itérations  
        Licence Informatique . Parcours : SDIA 
        Semestre 6  
        Module : Machine learning 
         
         
        Returns:  
           w (ndarray, shape (n,)) : paramètres finaux  
           b (scalar) : biais final  
           J_history (list) : historique du coût  
        """ 
 
        J_history = [] 
        w = copy.deepcopy(w_in) 
        b = b_in 
      
        for i in range(num_iters): 
            dj_db, dj_dw = gradient_function(X, y, w, b) 
            w = w - alpha * dj_dw 
            b = b - alpha * dj_db 
      
            if i < 100000: 
                J_history.append(cost_function(X, y, w, b)) 
            if i % math.ceil(num_iters / 10) == 0: 
                print(f"Iteration {i:4d} : Cout = {J_history[-1]:8.2f}") 
      
        return w, b, J_history 
  
# Execution 
initial_w = np.zeros_like(w_init) 
initial_b = 0. 
iterations = 1000
alpha = 5.0e-7
 
w_final, b_final, J_hist = gradient_descent( 
    X_train, y_train, initial_w, initial_b, 
    compute_cost, compute_gradient, alpha, iterations) 
  
print(f"b, w trouves : {b_final:.2f}, {w_final}") 
for i in range(X_train.shape[0]): 
    pred = np.dot(X_train[i], w_final) + b_final 
    print(f"Predit : {pred:.2f}  |  Reel : {y_train[i]}") 
fig, (ax1, ax2) = plt.subplots(1, 2, constrained_layout=True, figsize=(12, 4)) 
ax1.plot(J_hist) 
ax1.set_title("Cout vs. iterations (complet)"); ax1.set_xlabel("Iteration"); 
ax1.set_ylabel("Cout J") 
ax2.plot(100 + np.arange(len(J_hist[100:])), J_hist[100:]) 
ax2.set_title("Cout vs. iterations (fin)"); ax2.set_xlabel("Iteration"); 
ax2.set_ylabel("Cout J") 
plt.show() 

```

```python
def zscore_normalize_features(X): 
    """ 
    Normalisation Z-score de chaque feature (colonne) de X. 
    Args: 
        X (ndarray, shape (m,n)) : donnees d'entree 
    Returns: 
        X_norm  (ndarray, shape (m,n)) : donnees normalisees 
        mu      (ndarray, shape (n,))  : moyenne de chaque feature 
        sigma   (ndarray, shape (n,))  : ecart-type de chaque feature 
    """ 
    mu    = np.mean(X, axis=0)   # moyenne par colonne -> shape (n,) 
    sigma = np.std(X, axis=0)    # ecart-type par colonne -> shape (n,) 
    X_norm = (X - mu) / sigma    # normalisation element par element 
    return X_norm, mu, sigma 
```

```python
mu    = np.mean(X_train, axis=0) 
sigma = np.std(X_train, axis=0) 
 
X_mean = (X_train - mu)           # étape 1 : centrage 
X_norm = (X_train - mu) / sigma   # étape 2 : réduction 
 
fig, ax = plt.subplots(1, 3, figsize=(12, 3)) 
 
ax[0].scatter(X_train[:, 0], X_train[:, 3]) 
ax[0].set_title("Non normalisé") 
ax[0].set_xlabel("surface(sqft)"); ax[0].set_ylabel("age") 
ax[0].axis('equal') 
 
ax[1].scatter(X_mean[:, 0], X_mean[:, 3]) 
ax[1].set_title("Après centrage (X - μ)") 
ax[1].set_xlabel("surface(sqft)") 
ax[1].axis('equal') 
 
ax[2].scatter(X_norm[:, 0], X_norm[:, 3]) 
ax[2].set_title("Normalisé Z-score") 
ax[2].set_xlabel("surface(sqft)") 
ax[2].axis('equal') 
 
plt.suptitle("Distribution des features avant, pendant et après normalisation") 
plt.tight_layout() 
plt.show() 
```

```python
X_norm, X_mu, X_sigma = zscore_normalize_features(X_train) 
print(f"Moyenne par feature     : {X_mu}") 
print(f"Ecart-type par feature  : {X_sigma}") 
print(f"Plage brute  (max-min)  : {np.ptp(X_train, axis=0)}") 
print(f"Plage normalisee        : {np.ptp(X_norm,  axis=0)}") 
```
