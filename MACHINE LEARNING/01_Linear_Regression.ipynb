```python
def compute_cost(x, y, w, b):
        """
        Calcule la fonction de coût pour la régression linéaire.
        Args:
        x (ndarray (m,)) : données, m exemples
        y (ndarray (m,)) : valeurs cibles
        w, b (scalaire) : paramètres du modèle
        Returns:
        total_cost (float) : coût total J(w,b)
        """
        m = x.shape[0]
        cost_sum = 0
        for i in range(m):
            f_wb = w * x[i] + b
        cost = (f_wb - y[i]) ** 2
        cost_sum = cost_sum + cost
        total_cost = (1 / (2 * m)) * cost_sum
        return total_cost
```

```python
import math, copy
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('./deeplearning.mplstyle')
from lab_utils_uni import plt_house_x, plt_contour_wgrad, plt_divergence,plt_gradients
```

```python
# Chargement des données
x_train = np.array([1.0, 2.0]) # features
y_train = np.array([300.0, 500.0]) # valeurs cibles
```

```python
def compute_gradient(x, y, w, b):
        """
        Calcule le gradient pour la régression linéaire.
        Args:
        x (ndarray (m,)) : données, m exemples
        y (ndarray (m,)) : valeurs cibles
        w, b (scalaire) : paramètres du modèle
        Returns:
        dj_dw (scalaire) : gradient du coût par rapport à w
        dj_db (scalaire) : gradient du coût par rapport à b
        """
        m = x.shape[0]
        dj_dw = 0
        dj_db = 0
        for i in range(m):
            f_wb = w * x[i] + b
            dj_dw_i = (f_wb - y[i]) * x[i]
            dj_db_i = f_wb - y[i]
        dj_db += dj_db_i
        dj_dw += dj_dw_i
        dj_dw = dj_dw / m
        dj_db = dj_db / m
        return dj_dw, dj_db
```

```python
plt_gradients(x_train, y_train, compute_cost, compute_gradient)
plt.show()
```

```python
def gradient_descent(x, y, w_in, b_in, alpha, num_iters, cost_function, 
gradient_function): 
    """ 
    Effectue la descente de gradient pour trouver w, b optimaux. 
    Args: 
      x (ndarray (m,))  : données, m exemples 
      y (ndarray (m,))  : valeurs cibles 
      w_in, b_in (scalaire) : valeurs initiales des paramètres 
      alpha (float)     : taux d'apprentissage 
      num_iters (int)   : nombre d'itérations 
      cost_function     : fonction pour calculer le coût 
      gradient_function : fonction pour calculer le gradient 
    Returns: 
      w, b (scalaire) : paramètres optimisés 
      J_history (List): historique du coût 
      p_history (list): historique des paramètres [w,b] 
    """ 
    w = copy.deepcopy(w_in)  # évite de modifier w_in global 
    J_history = [] 
    p_history = [] 
    b = b_in 
    w = w_in 
    for i in range(num_iters): 
        # Calcul du gradient 
        dj_dw, dj_db = gradient_function(x, y, w, b) 
        # Mise à jour des paramètres 
        b = b - alpha * dj_db 
        w = w - alpha * dj_dw 
        # Sauvegarde de l'historique 
        if i < 100000:  # prévention d'épuisement des ressources 
            J_history.append(cost_function(x, y, w, b)) 
            p_history.append([w, b]) 
        # Affichage tous les 10e de num_iters 
        if i % math.ceil(num_iters / 10) == 0: 
            print(f"Iteration {i:4}: Coût {J_history[-1]:0.2e}, ", 
                  f"dj_dw: {dj_dw: 0.3e}, dj_db: {dj_db: 0.3e},", 
                  f"w: {w: 0.3e}, b: {b: 0.5e}") 
    return w, b, J_history, p_history 
```

```python
# Initialisation des paramètres 
w_init = 0 
b_init = 0 
# Paramètres de la descente 
iterations = 10000 
tmp_alpha = 1.0e-3 
# Exécution de la descente de gradient 
w_final, b_final, J_hist, p_hist = gradient_descent( 
    x_train, y_train, w_init, b_init, 
    tmp_alpha, iterations, 
    compute_cost, compute_gradient) 
print(f"(w,b) trouvés : ({w_final:8.4f}, {b_final:8.4f})") 
```
